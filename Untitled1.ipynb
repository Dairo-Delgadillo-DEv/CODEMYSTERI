Registro sobre Banca y Riesgo Crediticio

```python
## Importamaos librerias utiles
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Generación de datos sintéticos

```python
## Configuracion de tamaño
n_filas = 100000
```

```python
np.random.seed(42)
data = {
    'ID_Cliente': range(1, n_filas + 1),
    'Edad': np.random.randint(18, 95, n_filas),
    'Ingresos_Anuales': np.random.normal(35000, 15000, n_filas),
    'Puntuacion_Credito': np.random.randint(300, 850, n_filas),
    'Deuda_Total': np.random.uniform(0, 50000, n_filas),
    'Num_Productos': np.random.randint(1, 5, n_filas),
    'Ciudad': np.random.choice(['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Bilbao'], n_filas),
    'Es_Premium': np.random.choice([0, 1], n_filas, p=[0.8, 0.2]),
    'Gasto_Mensual_Promedio': np.random.normal(1200, 500, n_filas),
    'Churn': np.random.choice([0, 1], n_filas, p=[0.85, 0.15]) # Nuestra variable objetivo (Target)
    }

df = pd.DataFrame(data)

# --- INYECTANDO RUIDO PARA PRACTICAR EDA ---
# 1. Introducir valores nulos (NaN)
df.loc[df.sample(frac=0.05).index, 'Ingresos_Anuales'] = np.nan
df.loc[df.sample(frac=0.02).index, 'Edad'] = np.nan

# 2. Introducir Outliers (valores locos)
df.loc[0:10, 'Ingresos_Anuales'] = 5000000  # Multimillonarios
df.loc[11:20, 'Edad'] = -5                  # Errores de sistema
df.loc[21:30, 'Edad'] = 150                 # Edades imposibles

# 3. Introducir duplicados
df = pd.concat([df, df.iloc[:500]], ignore_index=True)

# Guardar
df.to_csv('banca_datos_robustos.csv', index=False)
print("¡Dataset de 100,500 filas generado con éxito como 'banca_datos_robustos.csv'!")
## --- Verificamos si existen datos Null con (isnull) ---
df.isnull().sum()
## --- Verificamos si existen datos duplicados ---
df.drop_duplicates()


```

Análisis Estadístico (Detección de Outliers)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configuramos el estilo
sns.set_theme(style="whitegrid")

# Creamos una figura con dos subgráficos (uno para Edad y otro para Ingresos)
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Boxplot para Edad
sns.boxplot(ax=axes[0], x=df['Edad'], color='lightcoral')
axes[0].set_title('Detección de Outliers en Edad')

# Boxplot para Ingresos
sns.boxplot(ax=axes[1], x=df['Ingresos_Anuales'], color='skyblue')
axes[1].set_title('Detección de Outliers en Ingresos')

plt.show()

```

```python
# Definimos un umbral (basado en lo que vimos en el boxplot)
umbral_riqueza = df['Ingresos_Anuales'].quantile(0.99) # Top 1%

# Creamos una variable categórica (Feature Engineering)
df['Perfil_Cliente'] = 'Normal'
df.loc[df['Ingresos_Anuales'] > umbral_riqueza, 'Perfil_Cliente'] = 'Ultra_High_Wealth'

print(df['Perfil_Cliente'].value_counts())
```

```python
# Función eficiente para limpiar outliers usando IQR
def limpiar_outliers(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Filtramos el dataframe
    return df[(df[columna] >= limite_inferior) & (df[columna] <= limite_superior)]

# Aplicamos la limpieza a la Edad (donde hay errores claros)
df_limpio = limpiar_outliers(df, 'Edad')

print(f"Filas antes: {len(df)} | Filas después: {len(df_limpio)}")
```

Correlación con el Target (Churn)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Aplicamos One-Hot Encoding a la columna 'Ciudad'
# Usamos drop_first=True para evitar la "trampa de la variable ficticia" (multicolinealidad)
df_preprocesado = pd.get_dummies(df, columns=['Ciudad'], prefix='City', drop_first=True)

# 2. Calculamos la matriz de correlación de todas las variables numéricas
# El método 'pearson' es el estándar para buscar relaciones lineales
matriz_corr = df_preprocesado.corr(numeric_only=True)

# 3. Visualizamos con un Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(matriz_corr,
            annot=True,       # Muestra los números de correlación
            fmt=".2f",        # Limita a 2 decimales
            cmap='coolwarm',  # Azul (negativa), Rojo (positiva)
            linewidths=0.5)

plt.title('Mapa de Calor de Correlación: Variables vs Churn')
plt.show()
```
